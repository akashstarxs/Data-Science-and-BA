The NIFTY 50 is a benchmark Indian stock market index that represents the weighted average of 50 of the largest Indian companies listed on the National Stock Exchange (NSE). These companies span various sectors of the Indian economy, making the NIFTY 50 a crucial indicator of market performance and investor sentiment. Despite its relatively small number of companies, the NIFTY 50 still represents over 60% of the free float market capitalization of the stocks listed on the NSE. It serves as a benchmark for fund portfolios, index funds, and derivatives trading, and includes companies from sectors such as IT, banking, energy, consumer goods, and more.

# **<font color='Green'> Visualization and Comparison**

```python
#Import required packages and Libraries
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  Select 10 stocks from NIFTY 50
selected_stocks = [
    "INFY.NS",  # Infosys Ltd : IT Services
    "LT.NS",    # Larsen & Toubro Ltd : Construction
    "SBILIFE.NS",  # SBI Life Insurance Company Ltd : Insurance
    "ADANIPORTS.NS",  # Adani Ports and Special Economic Zone Ltd : Transport Infra
    "RELIANCE.NS",  # Reliance Industries : Oil.Gas& Fuels
    "SBIN.NS",   # State Bank of India : Banks
    "HINDUNILVR.NS",  # Hindustan Unilever : Personal Products
    "BHARTIARTL.NS",  # Bharti Airtel : Telecom Services
    "M&M.NS",    # Mahindra & Mahindra Ltd : Auto Manufacturers
    "DRREDDY.NS" # Dr Reddy's Laboratories : Pharmaceutical
]
```

```python
# Price data for selected stocks over 5 years
start_date = "2019-02-17"
end_date = "2024-02-17"
stock_data = yf.download(selected_stocks, start=start_date, end=end_date)['Adj Close']
stock_data.tail(10)
```

```python
#Price Movement of Selected Stocks over 5 Years
plt.figure(figsize=(12, 8))

for stock in selected_stocks:
    plt.plot(stock_data.index, stock_data[stock], label=stock)

plt.title('Price Movement of Selected Stocks over 5 Years')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown', 'pink', 'gray', 'olive', 'cyan']
```

```python
# Plotting subplots for each stock with smaller subplots
fig, axes = plt.subplots(nrows=len(selected_stocks), ncols=1, figsize=(10, 15), sharex=True)

for i, stock in enumerate(selected_stocks):
    axes[i].plot(stock_data.index, stock_data[stock],color=colors[i % len(colors)], label=stock)
    axes[i].set_title(stock)
    axes[i].grid(True)
    axes[i].legend()

plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjusts the layout to prevent overlapping content
plt.show()
```

```python

window_size = 30

# Calculate the moving average for each stock
moving_averages = stock_data.rolling(window=window_size).mean()

# Plotting the moving averages
plt.figure(figsize=(12, 8))

for stock in selected_stocks:
    plt.plot(stock_data.index, stock_data[stock], label=stock)
    plt.plot(moving_averages.index, moving_averages[stock], label=f'{stock} Moving Average ({window_size} days)')

plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.title('Moving Averages of Selected Stocks')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python


# Calculate daily returns
daily_returns = stock_data.pct_change() * 100

# Drop the first row since it doesn't have a previous day's price for comparison
daily_returns = daily_returns.dropna()

print(daily_returns)
```

```python
# Plotting individual daily return plots for each stock
for i, stock in enumerate(selected_stocks):
    plt.figure(figsize=(8, 5))
    plt.plot(daily_returns.index, daily_returns[stock], color=colors[i % len(colors)], label=stock)
    plt.xlabel('Date')
    plt.ylabel('Daily Return')
    plt.title(f'Daily Returns of {stock}')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

# **<font color='Green'> Optimal portfolio weights**

```python


# Compute the average daily return for each stock
average_daily_returns = daily_returns.mean()

# Calculate the number of trading days in a year (typically 252 for stock markets)
trading_days_per_year = 252

# Annualize the average daily return for each stock
annual_returns = average_daily_returns * trading_days_per_year

from scipy.optimize import minimize

# Expected returns for each stock (annualized)
expected_returns = annual_returns.values

# Covariance matrix of returns
cov_matrix = daily_returns.cov().values

# Number of assets
num_assets = len(selected_stocks)

# Initial guess for portfolio weights (equal weights)
initial_weights = np.array([1 / num_assets] * num_assets)

# Constraints: sum of weights equals 1
constraints = ({'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1})

# Objective function: negative Sharpe Ratio (to maximize)
def negative_sharpe_ratio(weights, expected_returns, cov_matrix, risk_free_rate):
    portfolio_return = np.sum(expected_returns * weights)
    portfolio_std_dev = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    sharpe_ratio = (portfolio_return - risk_free_rate) / portfolio_std_dev
    return -sharpe_ratio

# Risk-free rate (you can set your own value)
risk_free_rate = 0.07

# Optimization to find optimal portfolio weights
result = minimize(negative_sharpe_ratio, initial_weights, args=(expected_returns, cov_matrix, risk_free_rate), method='SLSQP', constraints=constraints)

# Obtained portfolio weights
optimal_weights = result.x

print("Optimal Portfolio Weights:")
for stock, weight in zip(selected_stocks, optimal_weights):
    print(f"{stock}: {weight:.4f}")
```

```python
from tabulate import tabulate

# Create a list of tuples containing the stock symbols and their corresponding weights
portfolio_data = [(stock, weight) for stock, weight in zip(selected_stocks, optimal_weights)]

# Print the formatted table
print(tabulate(portfolio_data, headers=['Stock', 'Weight'], tablefmt='fancy_grid'))
```
